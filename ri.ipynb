```python
import numpy as np
import random
```

```python
n_arms=3
#we have 3 slots machines
```

```python
true_rewards= [0.2,0.5,0.8]
```

```python
success=np.ones(n_arms)
failure=np.ones(n_arms)
```

```python
episodes=1000
total_reward=0
```

```python
for episodes in range (episodes):
    #machine is guessing the best arm to pull
    sample_values=[
        #genertsing random probabilty baser don old success and failure rate 
        np.random.beta(success[i],failure[i])
        for i in range (n_arms)
    ]
    #pick the highest sample values
    action=np.argmax(sample_values)
    #if random probabilty is less than true value 
    reward=1 if random.random() < true_rewards[action] else 0

    if reward==1:
        success[action]+=1
    else:
        failure[action]+=1
```

```python
print("success ",success)
print('failure',failure)
print('total rewards',total_reward)
print('best arm learned',np.argmax(success))
```
