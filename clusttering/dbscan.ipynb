- DBSCAN- density based spatial clustering of application with noise 
   - outliners(noise) part of data but are not inculded in clusters 
   - to mkae the data more efficent 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('Mall_Customers.csv')
df
```

```python
X=df.iloc[:,[3,4]].values
```

```python
X=StandardScaler().fit_transform(X)
```

- eps= kneighbour hood raidus(how close two point must be in the same cluster )
- min_samples=minimum points needed to form a cluster 

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)#eps=epsilon
clusters=dbscan.fit_predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=clusters,cmap='rainbow')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('DBSCAN Clustering')
plt.show()
```

#finding outliers

```python
outliers=X[clusters==-1]# Identifying outliers -1 indicates noise points in DBSCAN
print("Outliers detected by DBSCAN:",len(outliers))
```

```python
plt.scatter(X[:,0],X[:,1],c=clusters,cmap='viridis',s=50)
plt.scatter(outliers[:,0],outliers[:,1],c='red',s=100,label='Outliers',edgecolors='k')
plt.title('DBSCAN Clustering with Outliers Highlighted')
```
