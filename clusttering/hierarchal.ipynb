- hierarchial clusttering
  - Agglomerative (bottom-up)which starts with individual points and merges them into larger clusters
  - Divisive (top-down) which starts with individual points and merges them into larger clusters  

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Mall_Customers.csv')
df
```

```python
x=df.iloc[:,[3,4]].values #age and income
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

##### in dendogram we see the maximum lenghth of line that isnt cut by another

```python
from sklearn.cluster import AgglomerativeClustering
hc= AgglomerativeClustering(n_clusters=5,metric='euclidean',linkage='ward')
y_hc=hc.fit_predict(x)
```

```python
plt.scatter(x[y_hc==0,0],x[y_hc==0,1],color='green',label='cluster 1')
plt.scatter(x[y_hc==1,0],x[y_hc==1,1],color='red',label='cluster 2')
plt.scatter(x[y_hc==2,0],x[y_hc==2,1],color='black',label='cluster 3')
plt.scatter(x[y_hc==3,0],x[y_hc==3,1],color='blue',label='cluster 4')
plt.scatter(x[y_hc==4,0],x[y_hc==4,1],color='yellow',label='cluster 5')

plt.title('Hierarchal Clustering')
plt.xlabel('Annual income ')
plt.ylabel('Spending score')
plt.legend()
plt.show()
```
