- hierarchial clusttering
  - Agglomerative (bottom-up)which starts with individual points and merges them into larger clusters
  - Divisive (top-down) which starts with individual points and merges them into larger clusters  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Mall_Customers.csv')
df
```

```python
x=df.iloc[:,[3,4]].values #age and income
```

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```
