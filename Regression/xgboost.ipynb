```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from xgboost import XGBClassifier
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.isnull().sum()
```

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0,
                                                    stratify=y)
```

```python
xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=8 random_state=0)
xgb.fit(X_train, y_train)
```

```python
y_pred = xgb.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("Train class Distribution:\n", y_train.value_counts(normalize=True))
print("Test class Distribution:\n", y_test.value_counts(normalize=True))
```

```python
# we are learning how to msnage imbalanced datasets in classification tasks
print("Feature Importances:\n", pd.Series(xgb.feature_importances_, index=X.columns).sort_values(ascending=False))
```
